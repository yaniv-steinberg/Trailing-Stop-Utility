```python
#return the percentage dsiffernece between two numbers
def percentage_difference(high, low):
    return ((high - low) / high) * 100
```

```python
#returns the amount of intervals per day accorfing to interval's length
def interval_length_to_per_day(interval_length, exclude_extended_hours = True):
    day_hours = 6.5 if exclude_extended_hours else 16
    print(f'interval_length_to_per_day  {int((day_hours * 60) // interval_length)}')
    return int((day_hours * 60) // interval_length)
    
        
```

```python
#returns the number of minutes (number) from a string indicating the number of minutes
def interval_length_str_to_num(interval_length):
    switcher = {
        '1min': 1,
        '5min': 5,
        '15min': 15,
        '30min': 30,
        '60min': 60
    }
    return switcher.get(interval_length)
```

```python
#returns the first index to check or the the length of the df if the desired index
#exceeds the length of the df
def calc_first_index(df_len, days_back, interval_length, exclude_extended_hours = True):
    interval_length = interval_length_str_to_num(interval_length)
    return min(df_len, interval_length_to_per_day(interval_length, exclude_extended_hours) * days_back)
```

```python
#returns yes iff the time of the date is in extended hours (before 09:30 or after 16:00)
def is_extended_hours(date):
    #time = int(df.index[0][11] + df.index[0][12] + df.index[0][14] + df.index[0][15])
    time = int(date[11] + date[12] + date[14] + date[15])
    return time < 930 or time > 1600
```

```python
def calc_max_trailing_lost(df, days_back, interval_length, exclude_extended_hours = True):
    max_trailing_lost = 0
    max_trailing_lost_low_date = 'NA'
    max_trailing_lost_high_date = 'NA'
    max_trailing_lost_high_price = 0
    max_trailing_lost_low_price = 0
    highest = 0.00001
    lowest = float('inf')
    highest_date = 'NA'
    lowest_date = 'NA'
    
    first_index = calc_first_index(len(df),days_back, interval_length)
    print(f'first index is: {first_index}')
    #iterates from the first index to check to the current date (index 0)
    for index in reversed(range(first_index)):
        cur_high = float(df['high'][index])
        cur_low = float(df['low'][index])
        cur_date = df.index[index]
        
        if exclude_extended_hours and is_extended_hours(cur_date):
            continue
        
        if cur_low < lowest:
            lowest = cur_low
            lowest_date = cur_date
        
            cur_max_trailing_lost = percentage_difference(highest, lowest)
            if cur_max_trailing_lost > max_trailing_lost:
                #print(f'{cur_max_trailing_lost}, {highest_date}, {lowest_date}')
                max_trailing_lost = cur_max_trailing_lost
                max_trailing_lost_low_date = lowest_date
                max_trailing_lost_high_date = highest_date
                max_trailing_lost_low_price = lowest
                max_trailing_lost_high_price = highest
        
        if cur_high > highest:
            highest = cur_high
            highest_date = cur_date
            lowest = cur_low
            lowest_date = cur_date
        
    return max_trailing_lost, max_trailing_lost_high_date, max_trailing_lost_high_price, max_trailing_lost_low_date, max_trailing_lost_low_price
```

```python
'''
Returns the slice at month_ago
'''
def get_slice(month_ago):
    bias_month_ago = month_ago - 1
    slice_year = bias_month_ago // 12 + 1
    slice_month = bias_month_ago - ((slice_year - 1) * 12)
    slice_name = 'year' + str(slice_year) + 'month' + str(slice_month + 1)
    return slice_name
```
